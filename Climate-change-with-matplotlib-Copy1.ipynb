# Climate Change with Matplotlib

## Climate Change and Global Warming:
 
The term “Climate change” includes global warming,but refers to the broader range of changes in Earth's climate system that result in new weather patterns that last for at least a few decades, and maybe for millions of years. “Global warming” refers to the long-term warming of the planet which most scientists agree is largely caused by human activity.

## Main Cause for Global Warming: Greenhouse Gas Emissions
Data Source: https://www.epa.gov/ghgemissions/global-greenhouse-gas-emissions-data 

```python
#Global Greenhouse gasses plot. 
data = [['C02 From Industry', 65], ['Methane', 16], ['C02 From Land Use', 11], ['Nitrous Oxide', 6],['F-gases', 2]]
colors_list = ['yellowgreen','gold', 'lightcoral', 'lightskyblue', 'pink']

df = pd.DataFrame(data, columns = ['Greenhouse Gas', 'Porcentage']) 

df['Porcentage'].plot(kind='pie', 
                        figsize=(20, 8),
                        autopct='%1.1f%%', 
                        startangle=90,    
                        shadow=True,       
                        labels=None,         
                        pctdistance=1.19,    
                        colors=colors_list)
                        

plt.title('Global Greenhouse Gas Emissions', y=1.12) 
plt.axis('equal') 
plt.legend(labels=df['Greenhouse Gas'], loc='upper left') 
plt.rc('font', size=20)

plt.show()
```

## Levels of carbon dioxide in the atmosphere from 1900 to 2019:
Data Source: https://www.climateinteractive.org/programs/scoreboard/scoreboard-science-and-data/ 

```python
#CO2 Global Emissions from 1900 to 2019 plot
FILE = 'global_CO2_emissions.txt'

with open(FILE, 'r') as fdin:
    list_of_data1 = [line.strip() for  line in fdin]

list_of_data2 = [float(data) for data in list_of_data1]
years = range(1900, 2020)

data_dictionary = dict(zip(years, list_of_data2))
x = list(data_dictionary.keys())
y = list(data_dictionary.values())

plt.plot(x, y, "r")
plt.title('CO2 Global Emissions from 1900 to 2019', fontsize=20, pad=30)
plt.xlabel('Years', fontsize=14)
plt.ylabel('Gigatons per year', fontsize=14)
plt.grid(True)
plt.rcParams['figure.figsize'] = [11, 8]

plt.show()
```

## Global surface temperature levels from 1880 to 2018:

Data Source: https://climate.nasa.gov/vital-signs/global-temperature/ </p>

```python
#Global Temperature from 1880 to 2018 plot.

FILE = 'global_temperature.txt'

with open(FILE, 'r') as fdin:
     list_of_data1 = [line.strip() for  line in fdin]

list_of_data2 = [float(data) for data in list_of_data1]
years = range(1880, 2019)

data_dictionary = dict(zip(years, list_of_data2))
x = list(data_dictionary.keys())
y = list(data_dictionary.values())

plt.bar(x, y, width=1, color="#D68910")
plt.xlabel("Years")
plt.ylabel("Temperature Anomaly (C)")
plt.title('Global Temperature', fontsize=20, pad=30)
plt.grid(True)

plt.show()
```
